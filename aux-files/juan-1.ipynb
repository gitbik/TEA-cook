```python
import pandas as pd
data = pd.DataFrame({"A" : [ 5, 23, 56, 7, 8],
                "B" : [ 2, 43, 52, 68, 6],
                "C" : [ 41, 5, 7, 2, 32],
                "D" : [ 14, 3, 16, 2, 36],
                "E" : [ 45, 23, 66, 32, 23]})

print("Original DataFrame :")
display(data)
def highlight_rows(x):
    if x.E>40:
        return['background-color: pink']*5
    else:
        return['background-color: blue']*5
print("Highlighted DataFrame :")
display(data.style.apply(highlight_rows, axis = 1))

dat1 = (data.style.apply(highlight_rows, axis = 1))
print(dat1)
```
